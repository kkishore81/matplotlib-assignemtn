```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

# Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
The subplot() function is used to create multiple plots in a single figure. This is useful when you want to compare different data sets or show different trends in the same data set.
```

```python
import numpy as np

```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns
fig, axes = plt.subplots(2, 2)

# Plot the first line plot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Line 1')

# Plot the second line plot
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Line 2')

# Plot the third line plot
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Line 3')

# Plot the fourth line plot
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Line 4')

# Show the figure
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
#A bar plot is a type of graph that is used to show the comparison of categorical data. The bars on the plot represent the values of the data, 
#and the length of each bar is proportional to the value that # it represents.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Plot a bar plot
plt.bar(company, profit)
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Plot a horizontal bar plot
plt.barh(profit, company)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
#A box plot is a type of graph that is used to show the distribution of data. It shows the minimum, first quartile, median, third quartile, and maximum values of the data.
#The box plot also shows the whiskers, which extend to show the rest of the distribution, except for points that are determined to be outliers.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure
fig = plt.figure()

# Add a box plot for box1
ax = fig.add_subplot(111)
bp = ax.boxplot([box1, box2])

# Add a title
plt.title('Box Plot')

# Add x-axis and y-axis labels
plt.xlabel('Data Set')
plt.ylabel('Value')

# Show the plot
plt.show()
```
